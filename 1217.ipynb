```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/weifong410180397/weifong410180397/refs/heads/main/0101_3.csv'

df1 = pd.read_csv(url, index_col=0, encoding='unicode_escape')
df1
```

```python
df1.columns
```

```python
# 先檢查每個欄位是否為字串，若是字串再移除逗號並轉為數字
def safe_convert(value):
    if isinstance(value, str):
        return float(value.replace(",", ""))
    else:
        return float(value)

# 計算公式
X1 = safe_convert(df1.loc['6031 LINE Bank', 'Total current tax assets'])
X2 = safe_convert(df1.loc['6031 LINE Bank', 'Total current tax liabilities'])
X3 = safe_convert(df1.loc['6031 LINE Bank', 'Total assets'])
X4 = safe_convert(df1.loc['6031 LINE Bank', 'Total retained earnings'])
X5 = safe_convert(df1.loc['6031 LINE Bank', 'Earnings Before Interest and Tax'])
X6 = safe_convert(df1.loc['6031 LINE Bank', 'Total equity'])
X7 = X3 - X6
X8 = safe_convert(df1.loc['6031 LINE Bank', 'Total comprehensive income'])


A = ( X1 - X2 ) / X3
B = X4 / X3
C = X5 / X3
D = X6 / X7
E = X8 / X3

Z_score_LB = 1.2*A + 1.4*B + 3.3*C + 0.6*D + 1.0*E
Z_score_LB
```
