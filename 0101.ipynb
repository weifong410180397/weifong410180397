```python
from fredapi import Fred
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fred = Fred(api_key='239c03dd8e5fe7b0dc5632e74e217dee')
```

```python
data_mich = fred.get_series('MICH')
data_unrate = fred.get_series('UNRATE')


df_mich = data_mich.to_frame(name="MICH")
df_unrate = data_unrate.to_frame(name="UNRATE")
df_merged = pd.merge(df_mich, df_unrate, left_index=True, right_index=True, how='inner')


plt.figure(figsize=(12, 6))
plt.plot(df_merged.index, df_merged['MICH'], label='MICH (預期通貨膨脹率)', color='blue')
plt.plot(df_merged.index, df_merged['UNRATE'], label='UNRATE (失業率)', color='red')
plt.title('預期通貨膨脹率 (MICH) 與失業率 (UNRATE)')
plt.xlabel('日期')
plt.ylabel('百分比')
plt.legend()
plt.grid()
plt.savefig('MICH_UNRATE_plot.jpg', format='jpeg')
plt.show()

corr_matrix = df_merged.corr()
sns.heatmap(corr_matrix, cmap="Blues", annot=True)
plt.title('相關係數熱圖')
plt.savefig('MICH_UNRATE_correlation.png', dpi=400)
plt.show()
```
